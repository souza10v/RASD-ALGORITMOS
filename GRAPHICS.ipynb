```python
class GRAPHICS:
      # CONSTRUTOR DA CLASSE GRAPHICS
      def __init__(self, MODEL, DATASET):
          """ 
          Class constructor
          
          Input: 
          MODEL: Contains the specifications of each model of chart according to the platform user manual, (Dictionary)
          DATASET: Dataset optimization procedure, (Dictionary)
          """
          self.__MODEL = MODEL
          self.__DATASET = DATASET
      # CONVERTEA ENTRADA EM SI PARA POLEGADAS    
      def CONVERT_SI_TO_INCHES(WIDTH, HEIGHT):
          """ 
          This function convert figure size SI to Inches
          
          Input: 
          WIDTH: Figure width in SI units, (float)
          HEIGHT: Figure height in SI units, (float)
          
          Output:
          WIDTH: Figure width in inches units, (float)
          HEIGHT: Figure height in inches units, (float)
          """
          WIDTH = WIDTH / 0.0254
          HEIGHT = HEIGHT / 0.0254
          return WIDTH, HEIGHT
      # SALVA A FIGURA NA PASTA OU CAMINHO DESEJADO
      def SAVE_GRAPHIC(NAME, EXT, DPI):
          """ 
          This function save graphics on a specific path
          Extensions options : 
          - 'svg'
          - 'png'
          - 'eps'
          - 'pdf'
          
          Input: 
          NAME: Path + Name figure, (string)
          EXT: file extension, (string)
          DPI: The resolution in dots per inch, (integer)
          """
          plt.savefig(NAME + EXT, dpi = DPI, bbox_inches="tight", transparent=True)
      # PLOTAGEM GRÁFICOS
      def PLOT(self):
          """ 
          This function plot and save graphics on a specific path
          
          Input: 
          MODEL: Contains the specifications of each model of chart according to the platform user manual, (Dictionary)
          DATASET: Dataset optimization procedure, (Dictionary)
  
          Obs.: See official documentation
          1: MODEL == 0 (2D plot OF and Fitness, https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)
          MODEL_INPUT = {'MODEL': 0,
                          'WIDTH': 0.40, 
                          'HEIGHT': 0.20, 
                          'NAME': 'WANDER', 
                          'DPI': 500, 
                          'EXTENSION': '.svg', 
                          'COLOR 1': '#DB3469',
                          'COLOR 2': '#3498DB',
                          'MARKER 1': 'o', 
                          'MARKER 2': 's',
                          'MARKER SIZE': 12,
                          'LINE WIDTH': 5,
                          'LINE STYLE': '-',
                          'OF LABEL': 'OF $kg^3$',
                          'GRID': 'ON'}        2: 
          2: MODEL == 1 (BLA BLA BLA)

          3: MODEL == 2 (BLA BLA BLA)

          4: MODEL == 3 (BLA BLA BLA)

          5: MODEL == 4 (BLA BLA BLA)
          """
          GRAPHIC_TYPE = self.__MODEL['MODEL']
          if GRAPHIC_TYPE == 0:
              GRAPHICS.PLOT_MODEL_1(self)

      # MÉTODO DE CADA GRÁFICO
      # MODELO DE LINHAS PARA FOT E FI EM UM ÚNICO LOCAL
      def PLOT_MODEL_1(self):
          # ATTRIBUTES
          NAME = self.__MODEL['NAME']
          EXT = self.__MODEL['EXTENSION']
          DPI = self.__MODEL['DPI']
          W = self.__MODEL['WIDTH'] 
          H = self.__MODEL['HEIGHT']
          COLOR_1 = self.__MODEL['COLOR 1']
          COLOR_2 = self.__MODEL['COLOR 2']
          MARKER_1 = self.__MODEL['MARKER 1']
          MARKER_2 = self.__MODEL['MARKER 2']
          MARKER_SIZE = self.__MODEL['MARKER SIZE']
          LINEWIDTH = self.__MODEL['LINE WIDTH']
          LINESTYLE = self.__MODEL['LINE STYLE']
          GRID = self.__MODEL['GRID']
          OF_DATA = self.__DATASET['OF']
          OF_LABEL = self.__MODEL['OF LABEL']
          FIT_DATA = self.__DATASET['FIT']
          # CONVERT UNITS
          W, H = GRAPHICS.CONVERT_SI_TO_INCHES(W, H)
          # FIGURE SIZE
          FIG, AX = plt.subplots(2, 1, figsize = (W, H), sharex = True)
          # LINE PLOT
          AX[0].plot(OF_DATA, marker = MARKER_1, color = COLOR_1, linestyle = LINESTYLE, linewidth = LINEWIDTH, markersize = MARKER_SIZE)
          AX[1].plot(FIT_DATA, marker = MARKER_2, color = COLOR_2, linestyle = LINESTYLE, linewidth = LINEWIDTH, markersize = MARKER_SIZE)
          # TICK
          AX[0].tick_params(axis = 'x', labelsize = 'large', which = 'both', width = 2, length = 7, rotation = 0, labelcolor = '#000000')
          AX[1].tick_params(axis = 'x', labelsize = 'large', which = 'both', width = 2, length = 7, rotation = 0, labelcolor = '#000000')
          #AX[1].tick_params(axis = 'x', labelsize = 'large', which = 'minor', width = 1.5, length = 1, rotation = 0, labelcolor = '#000000') https://matplotlib.org/stable/gallery/ticks_and_spines/major_minor_demo.html#sphx-glr-gallery-ticks-and-spines-major-minor-demo-py
          AX[0].tick_params(axis = 'y', labelsize = 'large', which = 'both', width = 2, length = 7, rotation = 0, labelcolor = '#000000')
          AX[1].tick_params(axis = 'y', labelsize = 'large', which = 'both', width = 2, length = 7, rotation = 0, labelcolor = '#000000')
          #AX[1].set_xticklabels(fontsize=50)
          #plt.xticks(fontsize=30)
          #plt.yticks(AX[0], fontsize=25)
          #plt.yticks(AX[1], fontsize=25)       
          # LABELS
          font = {'family': 'sans',
                  'color':  '#000000',
                  'weight': 'normal',
                  'size': 16}
          AX[1].set_xlabel('Iteration', fontdict = font)
          AX[0].set_ylabel(OF_LABEL, fontdict = font)
          AX[1].set_ylabel('Fitness', fontdict = font)
          # GRID
          if GRID == 'ON':
            AX[0].grid(color = 'grey', linestyle = '-.', linewidth = 1, alpha = 0.20)
            AX[1].grid(color = 'grey', linestyle = '-.', linewidth = 1, alpha = 0.20)
          GRAPHICS.SAVE_GRAPHIC(NAME, EXT, DPI)
          plt.show()
```
