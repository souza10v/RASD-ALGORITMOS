```python
from RASD_TOOLBOX import *
import numpy as np
import pandas as pd
```

```python
PX = ['NORMAL', 1, 0.15]
PY = ['NORMAL', 2000, 151]
PZ = ['NORMAL', 3, 0.3]
POP = [10, 100,  1000]

SETUP = {'N_REP': len(POP),
        'POP': POP,
        'N_G': 2,
        'D': 3,
        'VARS': [PX, PY, PZ],
        'MODEL': 'MCS'}
```

```python
def OBJ(X):
    P_X = X[0]
    P_Y = X[1]
    P_Z = X[2]
    
    D_0 = 3
    L = 100
    E = 30 * 10 ** 6
    W = 2
    T = 4

    # STATE LIMIT FUNCTION
    
    R_0 = P_Z
    S_0 = (4 * L ** 3 / (E * W * T)) * (((P_Y / T ** 2) ** 2  + (P_X / W ** 2) ** 2) ** 0.5)
    G_0 = - R_0 + S_0

    R_1 = P_Z * 1.10
    S_1 = ((4 * L ** 3 / (E * W * T)) * (((P_Y / T ** 2) ** 2  + (P_X / W ** 2) ** 2) ** 0.5)) * 1.05
    G_1 = - R_1 + S_1

    R = [R_0, R_1] 
    S = [S_0, S_1] 
    G = [G_0, G_1]


    return R, S, G
```

```python
RESULTS_TEST_1 = RASD_STOCHASTIC(SETUP, OBJ)
```

```python
DADOS_MCS_0 = RESULTS_TEST_1[0]['TOTAL RESULTS']
DADOS_MCS_1 = RESULTS_TEST_1[1]['TOTAL RESULTS']
DADOS_MCS_2 = RESULTS_TEST_1[2]['TOTAL RESULTS']
```
